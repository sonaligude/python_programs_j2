# conditional statements in python
1.python have 4 types of conditional statements
2.`if` statement
3. `elif` statement
4.`else` statement
5. `nested` statement

```python - if
if(condition): # if condition true inside statements are running
   statement1
   statements...
```

```python - if,else
if(condition):#if condition true inside statements are running
    statement1
    statements2...n
else: # if condition fails its went to else statement and print inside statements
    statement1
    statement2...n
```


``` python - if,elif,else
if (condition):
    statements
elif(condition):
    statements
elif (condition):
    statements
else :
     statements

```

````python
# write a program to find the given number id positive or negative or zero?
num = int(input("please enter number here:"))
if (num>0):
    print(num,"is positive number")
elif (num<0):
    print(num,"is negative number")
else:
    print(num,"is zero")
````

````python
"""
write a program to find the grades based on student marks
1. if marks in between 80 to 100 -A grade
2. if marks in between 60 to 79 - B grade
3. if marks in between 40 to 59 - C grade
4. if marks in between 25 to 39 - D grade
5. if marks in between 0 to 24 - Fail
6. if other than above marks output is - please enter correct marks
"""
marks = int(input("enter student marks here:"))
if (marks>=80 and marks<=100):
    print(marks,"A grade")
elif(marks>=60 and marks<=79):
                      print(marks,"B grade")
elif(marks>=40 and marks<=59):
    print(marks,"C grade")
elif(marks>=25 and marks<=39):
    print(marks,"D grade")
elif(marks>=0 and marks<=24):
    print(marks,"Fail")
else:
    print(marks,"please enter correct marks")
````

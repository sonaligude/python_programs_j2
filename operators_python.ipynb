# operators in python


## 6 types of operators in python
1. arithematic operators
2. assigment operators
3. comparision operators
4. logical operators
5. membership operators
6. identity operators


```python
# arithematic operators - +,-,*,/,//,%
a=10
b=20
c=20
d=30
print(a+b)
print(b-c)
print(b*d)
print(d/a)
print(d//a)
print(d%a)
```

```python
# assigment operators - +=,-=,*=,/=,//=,%=
a=100
a+=50
print(a)
```

```python
b=200
b-=400
print(b)
```

```python
c=50
c*=100
print(c)
```

```python
# comparision operators - >,<,>=,<=,==,!=
a=10
b=20
c=20
d=30
print(a>b)
print(b==c)
print(b<c)
print(b!=c)
print(b>=a)
print(b<=a)
```

```python
# logical operator - and,or,not
"""

and - both conditions are true then result is true ,rest case it false
or - either one condition is true the result is true, rest case result is false
and truth table
t t =t
t f =f
f t =f
f f =f
or truth table 
t t =t
t f =t
f t =t
f f =f
"""


a=10
b=20
c=20
d=30
print(a>b and b==c)
print(a<b and c==b)
print(a>b or b==c)
print(not a>b and b==c)
```

```python
# membership operator -  in,not in
# validating members are present in that sequence or group 
names= ["sonali","bhargavi","ananya","varshitha"]
print("sonali" in names)
print("bhargavi" not in names)
print("ananya" not in names)
print("varshitha" in names)
```

```python
# identity operators - is,is not
a=10000
b=10000
print(a is b)
print(id(a))
print(id(b))
```

```python
a=10
b=10
print(a is b)
print(id(a))
print(id(b))
```
